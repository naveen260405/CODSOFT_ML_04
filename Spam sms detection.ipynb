```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
```

```python
mail_data = pd.read_csv('/content/spam.csv')
mail_data.head()
```

```python
#Checking for the shape(number of rows and columns)
mail_data.shape
```

```python
#Counting the number of spam and ham messages by grouping it
mail_data.groupby('Category').describe()
```

```python
#label encoding spam mail as 0 and ham mail as 1
mail_data.loc[mail_data['Category'] == 'spam', 'Category'] = 0
mail_data.loc[mail_data['Category'] == 'ham', 'Category'] = 1
```

spam - 0

ham - 1

```python
#Separating the dataset as X(Message) and Y(Category) to get the independent(X) and dependent(Y) variables
X = mail_data['Message']
X
```

```python
Y = mail_data['Category']
Y
```

```python
#Splitting the dataset into training data and testing data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 50)
#Checking for dataset splitting
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
#Converting the text data into feature vectors(numerical values)
feature_vectors = TfidfVectorizer(min_df = 1, stop_words = 'english', lowercase=True)

X_train_features = feature_vectors.fit_transform(X_train)
X_test_features = feature_vectors.transform(X_test)
```

```python
#Converting Y_train and Y_test values as integers
Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
#Fitting the Multinomial naive bayes model
instance = MultinomialNB()
instance.fit(X_train_features, Y_train)
```

```python
prediction_on_training_data = instance.predict(X_train_features)
```

```python
#Checking for the accuracy score on training data
accuracy_of_training_data = accuracy_score(Y_train, prediction_on_training_data)
accuracy_of_training_data
```

```python
# instance.fit(X_test_features, Y_test)
prediction_on_testing_data = instance.predict(X_test_features)
```

```python
# #Checking for the accuracy score on testing data
accuracy_of_testing_data = accuracy_score(Y_test, prediction_on_testing_data)
accuracy_of_testing_data
```
